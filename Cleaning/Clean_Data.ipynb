# Formatando tabela Camp_Brasileiro_2024

```python
df_games = pd.read_csv("../Data/Camp_Brasileiro_2024.csv")
```

- Renomeando Colunas

```python
df_games = df_games.rename(columns={'venue.name': 'Stadium'})
df_games = df_games.rename(columns={'home.name': 'Home_Team'})
df_games = df_games.rename(columns={'away.name': 'Away_Team'})
df_games = df_games.rename(columns={'home': 'Gols_Home'})
df_games = df_games.rename(columns={'away': 'Gols_Away'})
df_games = df_games.rename(columns={'halftime.home': 'Half_Time_Gols_Home'})
df_games = df_games.rename(columns={'halftime.away': 'Half_Time_Gols_Away'})
```

1. Criando uma coluna booleana com os jogos que tiverem empate
2. Na coluna home.winner tem valores NaN, esses valores são de partidas que terminaram em empate. Vou colocar NaN como False

```python
df_games['Empate'] = df_games['Gols_Home'] == df_games['Gols_Away'].astype(int)
df_games['home.winner'] = df_games['home.winner'].fillna(False)
```

```python
df_games.to_csv("../Data_Lake/Camp_Brasileiro/2024/Camp_Brasileiro_2024_final")
```

# Formatando Camp_Brasileiro_team_stats

```python
df_team_stats = pd.read_csv("../Data/Camp_Brasileiro_2024_team_stats.csv")
```

```python
df_team_stats.isnull().sum()
```

```python
df_team_stats['red_cards'].fillna(0,inplace=True)
df_team_stats['yellow_cards'].fillna(0,inplace=True)
df_team_stats['offsides'].fillna(0,inplace=True)
df_team_stats['goals_prevented'].fillna(0,inplace=True)

df_team_stats.to_csv(("../Data_Lake/Camp_Brasileiro/2024/Camp_Brasileiro_2024_team_stats_final.csv"))
```
